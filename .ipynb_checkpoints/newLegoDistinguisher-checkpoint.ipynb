```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 


def dbeams(x, y):
    dx = x - 0
    dy = y - 0
    ans = dx**2 + dy**2
    result = m.sqrt(ans)
    return result


def dbrick(x, y):
    dx = x - 0
    dy = y - 0
    ans = dx**2 + dy**2
    result = m.sqrt(ans)
    return result

# Variable initialization
beamCounter = 0
brickCounter = 0

legos = np.loadtxt("Legos2.txt", delimiter=",")  # Importing the text file that contains the points.

beamlist = [(x, y) for x, y in legos if dbeams(x, y) < dbrick(x, y)]
bricklist = [(x, y) for x, y in legos if dbeams(x, y) > dbrick(x, y)]

for x, y in legos:  # Steps through the pairs of points
    if dbeams(x, y) < 0.8:  # Compares the distance
        beamCounter += 1
        plt.scatter(x, y, color="magenta")
    else:
        brickCounter += 1
        plt.scatter(x, y, color="turquoise")

print("Number of beams: %r" % beamCounter)
print("Number of bricks: %r" % brickCounter)
```
